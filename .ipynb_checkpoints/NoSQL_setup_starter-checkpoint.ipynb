1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
new_restaurant_data = {
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
result = establishments.insert_one(new_restaurant_data)
```

```python
# Check that the new restaurant was inserted
if result.acknowledged:
    print(f"Successfully inserted a new restaurant with ID: {result.inserted_id}")
else:
    print("Failed to insert the new restaurant.")
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Define the query to find the BusinessTypeID for "Restaurant/Cafe/Canteen"
query = {
    "BusinessType": "Restaurant/Cafe/Canteen"
}

# Define the projection to return only the BusinessTypeID and BusinessType fields
projection = {
    "_id": 0,  # Excludes the _id field
    "BusinessTypeID": 1,
    "BusinessType": 1
}

# Execute the query and projection
results = establishments.find(query, projection)

# Print
for result in results:
    print(result)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Define the query to find the document for "Penang Flavours"
query = {
    "BusinessName": "Penang Flavours"
}

# Define the update operation to set the BusinessTypeID
update = {
    "$set": {
        "BusinessTypeID":1  # Replace with the actual BusinessTypeID from above
    }
}

# Perform the operation
result = establishments.update_one(query, update)

# Print
if result.modified_count > 0:
    print("Successfully updated the BusinessTypeID for 'Penang Flavours'.")
else:
    print("No documents were updated.")
```

```python
# Confirm that the new restaurant was updated:
# Define the query to find the document for "Penang Flavours"
query = {
    "BusinessName": "Penang Flavours"
}

# Find the updated document
updated_document = establishments.find_one(query)

# Print
if updated_document:
    print("Updated Document:")
    print(updated_document)
else:
    print("Document not found.")
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
query = {
    "LocalAuthorityName": "Dover"
}

# Count the number of documents that match
count = establishments.count_documents(query)

# Print
print(f"Number of documents with LocalAuthorityName as 'Dover': {count}")
```

```python
# Delete all documents where LocalAuthorityName is "Dover":
query = {
    "LocalAuthorityName": "Dover"
}

# Delete all documents that match
result = establishments.delete_many(query)

# Print
print(f"Number of documents deleted: {result.deleted_count}")
```

```python
# Check if any remaining documents include Dover:
query = {
    "LocalAuthorityName": "Dover"
}

# Count the number of remaining documents that match
count = establishments.count_documents(query)

# Print
print(f"Number of remaining documents with LocalAuthorityName as 'Dover': {count}")
```

```python
# Check that other documents remain
remaining_document = establishments.find_one()

# Print
if remaining_document:
    print("At least one document remains in the collection:")
    print(remaining_document)
else:
    print("No documents found in the collection.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
query = {
    "geocode.longitude": {"$type": "string"},
    "geocode.latitude": {"$type": "string"}
}

# Define the update operation to set the data type to Decimal
update = {
    "$set": {
        "geocode.longitude": {"$toDouble": "$geocode.longitude"},
        "geocode.latitude": {"$toDouble": "$geocode.latitude"}
    }
}

# Apply the update operation
result = establishments.update_many(query, update)

# Print
print(f"Number of documents updated: {result.modified_count}")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
query = {
    "RatingValue": {"$type": "string"}
}

# Define the update operation to convert "RatingValue" to integer and set non-1-5 values to null
update = [
    {
        "$set": {
            "RatingValue": {
                "$cond": [
                    {"$in": ["$RatingValue", ["1", "2", "3", "4", "5"]]},
                    {"$toInt": "$RatingValue"},
                    None
                ]
            }
        }
    }
]

# Apply the update operation
result = establishments.update_many(query, update)

# Print
print(f"Number of documents updated: {result.modified_count}")
```

```python
# Change the data type from String to Integer for RatingValue
query = {
    "RatingValue": {"$type": "string"}
}

# Define the update operation to convert "RatingValue" to integer
update = {
    "$set": {
        "RatingValue": {"$toInt": "$RatingValue"}
    }
}

# Apply the update operation
result = establishments.update_many(query, update)

# Print
print(f"Number of documents updated: {result.modified_count}")
```
